```python
#import libraries
import pandas as pandas
import matplotlib
import pylab
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
pylab.rcParams['figure.figsize'] = (10., 10.)
```

```python
#set plot aesthetic!
plt.style.use('dark_background')
myColor = {'Brazil':'#2B2A4C'}
```

```python
#read in the dataset
df = pandas.read_csv('Brazilnewlocationdata.csv')
```

```python
#filter the dataframe
df['val_million'] = df['val']/1000000
df = df[df['measure']=='Prevalence']
df = df[df['age']=='All ages']
df = df[df['metric']=='Percent']
df['Number of Incidences of Diabetes (In Ten Thousands)']=df['val']/10000
df
```

```python
#plot the data!
sns.lineplot(data=df,x='year',y='Number of Incidences of Diabetes (In Ten Thousands)',hue='location').set_title('Number of Incidences of Type 2 Diabetes over Time in Brazil')
plt.legend(ncol=1,
           bbox_to_anchor=(1.01, 1), loc='upper left')
plt.xlabel('Year')
plt.show()
```
